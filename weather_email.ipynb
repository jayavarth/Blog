```python
import requests
import smtplib
from email.mime.text import MIMEText
from datetime import datetime, timedelta
import pytz
import time
import sched

API_KEY = "e10012f06d238d5574ba10a46cbaf2ed"
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

EMAIL = "jayavardhinimxia@gmail.com"
PASSWORD = "kcod zivm mnef potd"

scheduler = sched.scheduler(time.time, time.sleep)

def fetch_weather(location):
    params = {
        "q": location,
        "appid": API_KEY,
        "units": "metric"
    }
    response = requests.get(BASE_URL, params=params)
    if response.status_code == 200:
        data = response.json()
        return {
            "location": data["name"],
            "temperature": data["main"]["temp"],
            "humidity": data["main"]["humidity"],
            "pressure": data["main"]["pressure"],
            "weather": data["weather"][0]["description"],
            "high_temp": data["main"]["temp_max"],
            "low_temp": data["main"]["temp_min"],
            "wind_speed": data["wind"]["speed"],
            "visibility": data.get("visibility", "N/A") / 1000 if data.get("visibility") else "N/A",
            "sunrise": datetime.utcfromtimestamp(data["sys"]["sunrise"]).strftime('%I:%M %p'),
            "sunset": datetime.utcfromtimestamp(data["sys"]["sunset"]).strftime('%I:%M %p'),
            "cloud_coverage": data["clouds"]["all"]
        }
    else:
        print(f"Error fetching weather data: {response.status_code}")
        return None

def format_weather(weather):
    if weather:
        return f"""
        Weather Update for {weather['location']}:
        - Temperature: {weather['temperature']}°C
        - Humidity: {weather['humidity']}%
        - Condition: {weather['weather']}
        - High: {weather['high_temp']}°C | Low: {weather['low_temp']}°C
        - Wind Speed: {weather['wind_speed']} km/h
        - Pressure: {weather['pressure']} hPa
        - Visibility: {weather['visibility']} km
        - Sunrise: {weather['sunrise']}
        - Sunset: {weather['sunset']}
        - Cloud Coverage: {weather['cloud_coverage']}%
        """
    else:
        return "No weather data available."

def send_email(recipient, subject, body):
    try:
        msg = MIMEText(body)
        msg["Subject"] = subject
        msg["From"] = EMAIL
        msg["To"] = recipient

        with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
            server.login(EMAIL, PASSWORD)
            server.sendmail(EMAIL, recipient, msg.as_string())

        print(f"Email sent to {recipient}")
    except Exception as e:
        print(f"Error sending email: {e}")

def daily_weather_email_with_preferences(user_preferences):
    location = user_preferences["location"]
    recipient = user_preferences["email"]
    weather = fetch_weather(location)
    formatted_weather = format_weather(weather)
    send_email(recipient, "Today's Weather Update", formatted_weather)

def collect_user_preferences():
    user_location = input("Enter your location: ")
    user_email = input("Enter your email address: ")
    user_time = input("Enter the time for daily updates (e.g., 09:51 for 9:51 AM): ")
    return {
        "location": user_location,
        "email": user_email,
        "time": user_time
    }

def get_current_time():
    local_timezone = pytz.timezone("Asia/Kolkata")
    current_time = datetime.now(local_timezone).strftime("%H:%M")
    return current_time

def schedule_daily_email(user_preferences):
    user_time_str = user_preferences["time"]
    user_hour, user_minute = map(int, user_time_str.split(":"))

    current_time = datetime.now(pytz.timezone("Asia/Kolkata"))
    scheduled_time = current_time.replace(hour=user_hour, minute=user_minute, second=0, microsecond=0)

    if scheduled_time < current_time:
        scheduled_time += timedelta(days=1)
    delay_seconds = (scheduled_time - current_time).total_seconds()

    print(f"Scheduling email to be sent at {scheduled_time.strftime('%H:%M')}...")
    scheduler.enter(delay_seconds, 1, daily_weather_email_with_preferences, (user_preferences,))

def main():
    user_preferences = collect_user_preferences()
    schedule_daily_email(user_preferences)
    scheduler.run()

if __name__ == "__main__":
    main()
```
